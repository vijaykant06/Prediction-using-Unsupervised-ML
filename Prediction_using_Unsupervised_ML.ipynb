 <div class="alert alert-block alert-warning">
    <h3>Silhouette Width</h3>
    <p>The average silhouette approach measures the quality of a clustering. That is, it determines how well each object lies within its cluster. A high average silhouette width indicates a good clustering. The average silhouette method computes the average silhouette of observations for different values of k. The optimal number of clusters k is the one that maximizes the average silhouette over a range of possible values for k.</p>
</div> 

```r
#Silheuotte Width
sh = silhouette(k5$cluster,distance)
str(sh)
sil.summary=summary(sh)
sil.summary
```

 <div class="alert alert-block alert-success">

 <div class="alert alert-block alert-success">
    <h2>Optimal Number of cluster </h2>
</div> 

```r
# function to compute average silhouette for k clusters
avg_silouette = function(k) {
  km_result = kmeans(dat,k)
  ss = silhouette(km_result$cluster, distance)
  mean(ss[, 3])
}

# Compute and plot wss for k = 2 to k = 10
k.values = 2:10

# extract avg silhouette for 2-10 clusters
avg_sil_values = map_dbl(k.values, avg_silouette)

plot(k.values, avg_sil_values,
       type = "b", pch = 7, frame = FALSE, 
       xlab = "Number of clusters K",
       ylab = "Average Silhouettes")
```

 <div class="alert alert-block alert-success">
    <h3>Cluster marked with library Function</h3>
</div> 

```r
#Cluster marked with library Function
fviz_nbclust(dat, kmeans, method = "silhouette")
```

```r
#Elbow method
wss <- function(k) {
  kmeans(dat, k )$tot.withinss
}

# Compute and plot wss for k = 1 to k = 10
k.values <- 1:10

# extract wss for 1-10 clusters
wss_values <- map_dbl(k.values, wss)

plot(k.values, wss_values,
       type="b", pch = 19, frame = FALSE, 
       xlab="Number of clusters K",
       ylab="Total within-clusters sum of squares")
```

```r
fviz_nbclust(dat, kmeans, method = "wss")
```
